# 리스트

## **리스트 표현식(List comprehension)**
 - [표현식 for 변수 in 리스트]
 - 리스트 초기화시 사용, 특히 2차원 리스트 초기화시 효과적
 - 대괄호 안에 조건문, 반복문등을 적용해 리스트값 할당
 - ex. arr = [[0]*m for _ in range(n)] #좋은 예시
 - cf. arr = [[0]*m] *n #에러

```python
#0~9 까지
arr = [x for x in range(10)]
arr
```

```python
#0~19까지 홀수만 포함하는 수
arr = [x for x in range(20) if x%2 == 1] #반복문에 대한 필터링
#      (3)   ________(1)_____  ____(2)___ #표현식 처리 순서
arr
```

```python
#1~9까지 수들의 제곱값을 포함하는 리스트
arr = [x**2 for x in range(1, 10)]
arr
```

```python
# 4행 3렬 2차원 리스트 생성
r = 4
c = 3
arr = [['가']*r for _ in range(c)] # ['가', '가', '가', '가'] 각행마다 레퍼런스가 다름
print(arr)

arr[1][1] = 'A'
print(arr)
```

```python
#4행 3렬 2차원 리스트 잘못된 생성방법
r = 4
c = 3
arr2 = [['나']*r]*c # 같은 레퍼런스를 가지는 ['나', '나', '나', '나'] 리스트르 3행 만드는것
print(arr2)
arr2[1][1] = 'B'
print(arr2)
```

## 리스트 함수


```python
#리스트 선언
a = [1,2,3,4,5]
print('기본리스트:', a)

#리스트 추가
a.append(6)
print('삽입:', a)

#특정 인덱스에 데이터 추가
a.insert(2,0)
print('특정 인덱스에 데이터 추가:', a)

#리스트 오름차순정렬
a.sort()
print('오름차순정렬:', a)

#리스트 내림차순정렬
a.sort(reverse=True) #a의 메소드
print('내림차순정렬:', a)

#새롭게 정렬한 리스트 반환
b = sorted(a) #파이썬의 내장함수
print('새롭게 정렬한 리스트:', b)
print('기존 리스트:', a)
```

```python
a = [4,3,2,1]
#리스트 뒤집기
a.reverse()
print('리스트 뒤집기:', a)

#특정값의 데이터 개수 
a.append(4)
print('데이터 4추가:', a)
print('값이 4인 데이터 개수:', a.count(4))

#특정값 삭제 : 여러개가 있다면 앞에서 첫번째것을 삭제한다
a.remove(4)
print('특정값 삭제:', a)

#여러개의 값을 지워야할 경우 : remove all of 4 관련 함수가 없으므로 만들어야함
a= [1,2,3,4,5,5,5]
remove_set = {3,5} # remove_set에 지울 요소 값을 설정
new_a = [i for i in a if i not in remove_set]
print(new_a)

#리스트 삭제
del a[0:2]
del a[2]
print('리스트 삭제:',a)

#값에 대한 인덱스 반환
print('값에 대한 인덱스:', a.index(4))
```

# Dict (딕셔너리)
- 키와 값의 쌍을 데이터로 가지는 데이터
- 해시테이블 구조를 내부적으로 사용하므로 검색 및 수정에서 O(1)의 시간에 처리 가능

```python
data = dict()
data['a'] = 'apple'
data['b'] = 'banana'
data['c'] = 'cherry'

print(data)

if 'a' in data:
    print('apple exist in this data')
```

```python
print(data.keys())
print(data.values())
```

```python
print(list(data.keys()))
print(list(data.values()))
```

# 집합자료형(Set)
- 중복을 허용하지 않는 자료형
- 순서가 없음
- 순서가 없으므로 인덱싱으로 값을 얻을수 없음
- 해당 값이 있는지 없는지 확인 가능
- 중괄호 {} 안에 각원소를 콤마를 기준으로 넣어 초기화한다
- 집합연산
 - 합집합 : a|b
 - 교집합 : a&b
 - 차집합 : a-b

```python
data = set([1,1,1,2,3,4,4,5])
print(data)

data2 = {5,4,3,3,2,1,2}
print(data2)

if 5 in data2:
    print('5 is exist')
```

```python
#집합과 리스트의 연산시간 
a = [1,2,3,4,5]
print(3 in a) #O(N)

b = {1,2,3,4,5}
print(3 in b) #O(1) 상수시간 소요
```

```python
data = {1,2,3}

#원소 추가
data.add(4)
data.add(5)
print(data)

#여러 원소 추가
data.update([5,6])
print(data)

#특정 원소 삭제
data.remove(3)
print(data)
```

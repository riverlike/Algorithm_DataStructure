# 문제 :거스름 돈 

당신은 점원이다. 카운터에는 거스름돈으로 사용할 500원, 100원, 50원, 10원짜리 동전이 무한히 존재한다. 손님에게 거슬러줄 돈이 N원 일때 거슬러줄 동전의 최소 개수를 구하세요. 단, 거슬러 줄돈N은 항상 10의 배수입니다.

 - 문제해결 아이디어: 

가장큰화폐단위부터 돈을 거슬러주면된다. N=1260원일때
가장먼저 500원 2개, 100원 2개, 50원 1개, 10원 1개를 거슬러주면된다. 

 - 정당성 분석 : 

가장큰화폐단위부터 거슬러주는것이 최적의 해를 보장하는가? 
이문제는 큰단위(1000원)가 항상 작든단위들의(500, 50, 10)의 배수이므로 다른동전을 최소개수로 대체할 수 있기 때문이다.

- 시간복잡도 : O(N)

동전의 총 종류에만 영향을 받음

- 최적의 해가 되지 못는하는 경우 

800원을 거슬러주는 화폐단위가 500원, 400원 100원이라면?
큰단위부터 거슬러주는 방법으로 최적의 해가 나오지 않는다. 400원을 2개 주는것이 낳기때문인데 이런문제는 다이나믹프로그래밍으로 풀어야한다.

```python
n = 1260
count = 0

arr = [500,100,50,10]
for coin in arr:
    count += n//coin
    n = n%coin  #n %=coin

print(count)    
```

# 문제 :1이 될때까지

1. N에서 1을 뺀다.
2. N을 K로 나눈다.

N과 K가 주어질때 N이 1이 될때까지 1,2과정을 수행해야하는 최소회수를 구하시오.

- N이 17, K가 4인경우 1번과정을 1번 수행하면 N은 16이되고 2번과정 두번하면 1이된다. 전체과정을 실행한횟수는 3이된다.

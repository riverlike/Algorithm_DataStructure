# 반복문
- 반복적으로 실행하고자 할 때 사용하는 문법
- while : 조건식이 만족할때까지 반복 
- for : 특정범위까지 반복하거나 자료형(리스트, 튜플, 딕셔너리)을 돌면서 반복
 - range(시작값, 끝값, 스텝수) : 시작값부터 끝값 이전수까지 스텝만큰 건너뛰어서 반복
 



```python
#1부터 9까지의 값을 순회
for i in range(1, 10):
    print(i , end=' ')
```

```python
#1부터 9까지의 홀수값을 순회
for i in range(1, 10, 2):
    print(i , end=' ')
```

```python
#합격여부를 출력하되 컨닝을 한 학생을 제외하여 출력
scores = [90, 70, 80, 99, 100, 77, 88]
cheating_student = {3, 4 } #3번, 4번 학생
for i in range(len(scores)):
    if i+1 in cheating_student:
        continue
    if scores[i] >=80 :
        print(i+1, '번 : 합격')
```

```python
#구구단
for i in range(2, 10):
    for j in range(1, 10):
        print(i, 'x', j, '=', i*j)
    print()
```

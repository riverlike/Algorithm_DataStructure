# 순열과 조합
- 파이썬 기본 라이브러리에 지원
- 순열 
 - 서로 다른 n개에서 서로 다른 r개를 선택해 나열하는것
 - {a,b,c}에서 세개를 선택할 경우 : abc, acb, bac, bca, cab, cba
  - 3 * 2 * 1 = 6가지

 $$ {n} P {r} = n*(n-1)* \dots *(n-r +1) $$

- 조합
 - 서로 다른 n개에서 순서에 상관없이 서로 다른 r개를 선택해 나열하는것
 - {a,b,c}에서 순서를 고려하지않고 2개를 뽑을 경우 : ab,ac,bc 
  - 6/2! = 3 가지

  $$ {n} C {r} = \frac{n*(n-1)* \dots *(n-r+1)}{r!}  $$

```python
#순열 구하기
from itertools import permutations
data = [3,6,9]
res = list(permutations(data, 2))
print(res)
```

```python
#조합 구하기
from itertools import combinations
data = [3,6,9]
res = list(combinations(data, 2))
print(res)
```

# 중복순열과 중복조합
- 중복순열
 - 중복을 허용하여 모든 순열 구하는 경우
- 중복조합
 - 중복을 허용하여 모든 조합 구하는 경우

```python
#중복순열 구하기
from itertools import product
data = [3,6,9]
res = list(product(data, repeat=2)) #2개를 뽑는 모든 순열
print(res)
```

```python
#중복조합 구하기
from itertools import combinations_with_replacement
data = [3,6,9]
res = list(combinations_with_replacement(data, 2))
print(res)
```

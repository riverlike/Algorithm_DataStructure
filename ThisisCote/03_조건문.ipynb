# 조건문
- 프로그램의 흐름을 제어하는 문법
- 조건에 따라 실행 방식을 다르게 설정 할 수 있다.
- elif, else부분은 경우에 따라 사용하지 않음
- 들여쓰기 : 4개의 공백사용하는것이 표준

```python
score = 85

if score>=90:
    print('A')
elif score>=80:
    print('B')
elif score>=70:
    print('C')
elif score>=60:
    print('D')
else:
    print('F')
```

# 비교 연산자
-종류 : ==, != , > , < , >= , <=


# 논리 연산자
논리값 사이의 연산을 수행할 때 사용
- 종류 : and, or, not

```python
a = 1
if a>=0 and a<=10 :
    print('a는 0 이상 10이하의 수입니다.')
```

# 기타 연산자
- x in 리스트 : 리스트안에 x가 들어있을때 True 반환
- x not in 리스트 : 리스트안에 x가 들어있지 않을때 True 반환
- 리스트, 튜플, 문자열, 딕셔너리에서 사용가능

# 조건문의 간소화
- 조건문에서 실행할 코드가 한줄일때 줄바꿈 없이 간략히 표현가능
- 조건부 표현식은 if else문을 한줄에 작성하게 해줌

```python
score = 85

if score>=80:result='Pass'
else:result='Fail'

result
```

```python
score = 85

result = 'Pass' if score>=80 else 'Fail'
#        _____   ___________       _______
#     조건문 앞쪽이 True일때 실행, 뒷쪽은 False일때 실행됨
result
```

# 함수(Function)
- 특정한 작업을 하나의 단위로 묶어 높은것
- 불필요한 소스코드의 반복을 피하고 재사용성을 늘릴수 있다
- global 키워드
 - 함수 안에서 외부의 전역변수를 참조하기위해 global을 사용


```python
#매개 변수의 순서 없이 설정
def add(a,b):
    return a+b

print(add(b=3, a=1))
```

```python
# 함수안에서 전역변수를 사용할때 문제
a = 10

def func():
    a += 7

func()
```

```python
# 함수안에서 global 키워드 명시후 전역변수를 사용
a = 10

def func():
    global a
    a += 10

func()
print(a)
```

```python
#cf. 리스트 전역변수는 global 키워드없이 사용가능하다
a = [1,2,3]
def func():
    a.append(4)

func()
print(a)
```

# 람다
- 익명함수라고도 부름
- 특정한 기능을 수행하는 함수를 한줄에 작성 가능
- 여러번 재사용되지않고 한번쓰이고 끝날경우 사용
 - 내장 함수에서 자주 사용됨
 - ex. sorted 함수는 정렬기준에대한 파라마터로 함수를 설정할 수있다


```python
print( (lambda a,b:a+b)(3,5))
```

```python
# 점수기준 오름차순 정렬
arr = [('jenny', 90), ('son', 87), ('rm', 65),('mcu', 95)]

def my_key(x):
    return x[1] #튜플 인덱스
print(sorted(arr, key=my_key)) 
print(sorted(arr, key=lambda x:x[1])) 

# 이름순 정렬
print(sorted(arr, key=lambda x:x[0]))  #x는 arr의 한 원소
```

```python
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]

res = list(map(lambda a,b : a+b, list1, list2))
print(res)
```

# 자주 사용되는 내장함수

```python
# sum
print( sum([1,2,3,4,5]) )
```

```python
# min, max
res_min = min(4,6,2,1)
res_max = max(4,6,2,1)
print(res_min, res_max)
```

```python
# eval
print( eval('(3+5)*7'))
```

```python
# sorted
res = sorted([3,6,2,5,8])
reverse_res = sorted([3,6,2,5,8], reverse=True)
print(res)
print(reverse_res)
```

```python
# sorted with key
arr = [('jenny', 90), ('son', 87), ('rm', 65),('mcu', 95)]

print(sorted(arr, key=lambda x:x[1])) 
```
